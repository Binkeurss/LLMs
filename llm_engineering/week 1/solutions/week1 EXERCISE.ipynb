```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

# MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
openai = OpenAI(
    base_url="http://localhost:11434/v1",  # Trỏ về server local của Ollama
    api_key="ollama"                       # Key giả
)
```

```python
# here is the question; type over this to ask something new

# question = """
# Please explain what this code does and why:
# yield from {book.get("author") for book in books if book.get("author")}
# """

question_system_prompt = """
You are provided with a list of technical tools that provided by users.
You are able to create technical questions and explainations for each questions about the tools that would be most relevant,
such as questions for a use case, or a how to use, or problems when using them.
Respond in markdown without code blocks. as in this example:

{
    Lists
        Question: What is the NestJs
        Explanation: This is a framework for building efficient, scalable Node.js server-side applications. It uses TypeScript and incorporates elements of OOP (Object Oriented Programming), FP (Functional Programming), and FRP (Functional Reactive Programming).
        Question: Why do they not use NextJs to build backend applications?
        Explanation: In general, Next.js is a React framework primarily designed for building frontend applications, particularly for server-side rendering and static site generation. While it can be used to build backend applications, it is not optimized for that purpose. NestJS, on the other hand, is specifically designed for building scalable and maintainable server-side applications, making it a more suitable choice for backend development.
}
"""
```

```python
def get_tools_user_prompt(tools):
    user_prompt = f"""
Here is the list of tools that will make technical questions {tools} -
Please make technical questions and explanations for each question about the tools that would be most relevant,
such as questions for a use case, or a how to use, or problems when using them.

"""
    return user_prompt
```

```python
print(get_tools_user_prompt(["NestJs", "NextJs"]))
```

```python
# Get gpt-4o-mini to answer, with streaming
```

```python
# Get Llama 3.2 to answer

def stream_questions_tools(tools):
    stream = openai.chat.completions.create(
        # model="gpt-4.1-mini",
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": question_system_prompt},
            {"role": "user", "content": get_tools_user_prompt(tools=tools)}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
stream_questions_tools(["NestJs", "NextJs"])
```
